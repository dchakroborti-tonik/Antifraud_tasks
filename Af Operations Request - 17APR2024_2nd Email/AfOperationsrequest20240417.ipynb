```python
a = tuple(d['PID'])
b = tuple(d['TSA'])
print(f"a = \t{a[:5]}")
print(f"b - \t{b[:5]}")
```

```python
sq = f"""
with ldata as 
(select customerId, city, province, barangay, postalCode, custLoanJourneyCity,custLoanJourneyProvince, custLoanJourneyZipCode , natureofwork, industryDescription, docNumber, docType
, row_number() over (partition by customerId order by startApplyDateTime desc) rnk
from `risk_credit_mis.loan_master_table`)
select  
ca.OFCUSTOMERID PID,
ca.OFSTANDARDACCOUNTID TSA,
pd.FIRSTNAME||''||pd.MIDDLENAME||''||LASTNAME Name,
date(pd.DOB) DOB,
ca.OFDATEOPENED,
tcm.mobile_no mobileno,
coalesce(tf.House_Number)House_Number,
coalesce(tf.City_Municipality, ld.city) tsa_city,
coalesce(tf.Province,ld.province) tsa_province,
coalesce(tf.Barangay, ld.barangay) tsa_barangay,
coalesce(cast(tf.Zipcode as string) ,cast(ld.postalCode as string)) tsa_postalcode,
ld.custLoanJourneyCity loan_city,
ld.custLoanJourneyProvince loan_province,
ld.custLoanJourneyZipCode loan_postalcode,
coalesce(tf.Nature_Of_Work, ld.natureofwork) natureofwork,
coalesce(tf.Industry,ld.industryDescription) industry,
pd.EMPLOYERNAME,
tf.Document_Number,
tf.Document_Scan_Type document_type,
SEGCATEGORY KYC,
Device_Brand devicename
from prj-prod-dataplatform.core_raw.customer_accounts ca
left join prj-prod-dataplatform.dl_customers_db_raw.tdbk_customer_mtb tcm on ca.OFCUSTOMERID = tcm.cust_id
left join prj-prod-dataplatform.core_raw.party_details pd on pd.partyid = cast(ca.OFCUSTOMERID as numeric)
left join ldata ld on ld.customerId = cast(ca.OFCUSTOMERID as numeric)
left join prj-prod-dataplatform.marketing.tsa_funnel tf on tf.Customerid = cast(ca.OFCUSTOMERID as numeric)
where pd.PTYROLE = 'CUSTOMER'
and OFCUSTOMERID in {a}
and OFSTANDARDACCOUNTID in {b}
;
"""
print(sq)
```
